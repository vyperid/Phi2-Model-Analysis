```python
from langchain import PromptTemplate, LLMChain
```

```python
tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-2") #With tokenizer, our text prompt will become a format that model can understand.
```

```python
quantization_config = BitsAndBytesConfig(llm_int8_enable_fp32_cpu_offload=True) #Here with quantization we maintain the performance of the model while
#reduce the memory usage as well as computational requirements of the it. Here as a parameter we set llm_int8_enable_fp32_cpu_offload true. I learned that
#int8 means representing weights and activations using 8 bit integer values other than standard 32 bit values. This helps us reduce the memory usage and
#helps for faster computation. Also setting fp32_cpu_offload to true means that some of the computations will be getting executed in CPU with 32 bit
#floating point precision.
```

```python
import torch
```

```python
model = AutoModelForCausalLM.from_pretrained('microsoft/phi-2',
                                          load_in_8bit=True, #to tell the model to load weights in 8 bits
                                          torch_dtype = torch.float32, #model weights will be stored as 32 bit floating point numbers
                                          device_map  = 'auto', #selecting cpu or gpu automatically
                                          quantization_config=quantization_config,
                                          trust_remote_code = True #with this we say that we can trust the model and the codes (phi-2 in my case) because with
                                             #the model we also load the pre-trained weights
                                          )
```

```python
#creating a pipeline. throughout this learning process i've learned that pipelines are APIs that help us perform NLPs easier. First we set up the pipeline as
#"text generation", then upload the model and set some parameters. max_length is for max token numbers which is for controlling the max length of the output.
#temperature is about the quality of the output result. less temperature values will result in a more deterministic output while higher temperature values increase
#randomness. I saw that most of the people set this value to 0.3 and i applied the same. top_p plays with the probability of words that are getting used while sampling
#i set this value the same as the resource i followed while coding. repetition_penalty controls the repetitiveness of the output words or the tokens.
pipe = pipeline("text-generation",
               model=model,
               tokenizer=tokenizer,
               max_length=250,
               temperature=0.3,
               top_p=0.95,
               repetition_penalty=1.2)
local_llm = HuggingFacePipeline(pipeline=pipe)
pipe.model.config.pad_token_id = pipe.model.config.eos_token_id
```

```python
#for creating a template, {instruction} is for getting the specific prompt.
template = """Respond to the instruction below and behave like a chatbot then respond to the user. Do your best and try to be helpful.
### Instruction:

{instruction}

Answer:"""

prompt = PromptTemplate(template=template, input_variables=['instruction'])
```

```python
 #I created this function to use it with Flask.

 def output(prompt, local_llm, question):

  #creating a LLMChain object.
  llm_chain = LLMChain(prompt=prompt, llm=local_llm)

  return llm_chain.run(question)
```
